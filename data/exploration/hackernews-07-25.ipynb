```python
sorted_df = sorted_df[sorted_df.title != '']
```

```python
most_recent = sorted_df.groupby('title').agg({'create_time': max}).reset_index()
```

```python
sorted_df.merge(most_recent, 'inner', on=['title', 'create_time'])
```

```python
# Count number of times the article has appeared on the front page 
sorted_df.groupby('title').count()[['url']].reset_index()
```

```python
most_recent = (sorted_df.groupby('title')
                        .agg({'create_time': max, 'url': len})
                        .reset_index()
                        .rename(columns={'url': 'fp_count'}))
```

```python
most_recent
```

```python
sorted_df.merge(most_recent, 'inner', on=['title', 'create_time'])
```

```python
most_recent = (sorted_df.groupby('title')
                        .agg({'create_time': max, 'url': len, 'points': max})
                        .reset_index()
                        .rename(columns={'url': 'fp_count'}))
```

```python
tdf = sorted_df.merge(most_recent, 'inner', on=['title', 'create_time'])
```
